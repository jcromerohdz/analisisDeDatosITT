```python
pos=nx.spring_layout(g ) # Lo grafica igual 
nx.draw(g,pos,with_labels=True)  # Lo grafica
plt.show() #Lo muestra
```

```python
pos=nx.circular_layout(g ) # Lo grafica igual  de forma circular
nx.draw(g,pos, with_labels=True)  # Lo grafica
plt.show() #Lo muestra
```

```python
pos=nx.random_layout(g ) # Lo grafica igual  de forma circular
nx.draw(g,pos)  # Lo grafica
plt.show() #Lo muestra
```

```python
pos=nx.spectral_layout(g ) # Lo grafica igual  de forma circular
nx.draw(g,pos,with_labels=True)  # Lo grafica
plt.show() #Lo muestra
```

```python
pos=nx.spring_layout(g ) # Lo grafica igual  de forma circular
nx.draw(g,pos,with_labels=True)  # Lo grafica
plt.show() #Lo muestra
```

```python
pos=nx.spring_layout(g ) # Lo grafica igual  de forma circular
nx.draw(g,pos,with_labels=True,node_size=2)  # Lo grafica delimita el tamao del nodo en la grafica
plt.show() #Lo muestra
```

```python
pos=nx.spring_layout(g ) # Lo grafica igual  de forma circular
nx.draw(g,pos,with_labels=True,node_size=2, width = 5)  # Lo grafica delimita el tamao del nodo en la grafica
plt.show() #Lo muestra
```

```python
pos=nx.spring_layout(g ) # Lo grafica igual  de forma circular
nx.draw(g,pos,with_labels=True,node_size=[g.degree(i)*50 for i in g.nodes()], width = 2)  # Lo grafica delimita el tamao del nodo en la grafica
plt.show() #Lo muestra
```

```python
pos=nx.spring_layout(g ) # Lo grafica igual  de forma circular
nx.draw(g,pos,with_labels=True,node_size=[g.degree(i)*50 for i in g.nodes()], width = [1 if g.degree(i)<= 5 else 3 for i in g.nodes()])  # Lo grafica delimita el tamao del nodo en la grafica
plt.show() #Lo muestra
```

```python
pos=nx.spring_layout(g ) # Lo grafica igual  de forma circular
nx.draw(g,pos,with_labels=True,node_size=[g.degree(i)*50 for i in g.nodes()], width = [1 if g.degree(i)<= 5 else 3 for i in g.nodes()], node_color=[float(g.degree(i)) for i in g.nodes()])  # Lo grafica delimita el tamao del nodo en la grafica
plt.show() #Lo muestra
```

```python
pos=nx.spring_layout(g ) # Lo grafica igual  de forma circular
nx.draw(g,pos,with_labels=True,node_size=[g.degree(i)*50 for i in g.nodes()], width = [1 if g.degree(i)<= 5 else 3 for i in g.nodes()], node_color=[float(g.degree(i)) for i in g.nodes()],node_shapes='x')  # Lo grafica delimita el tamao del nodo en la grafica
plt.show() #Lo muestra
```

```python
pos=nx.spring_layout(g ) # Lo grafica igual  de forma circular
nx.draw(g,pos,with_labels=True,node_size=[g.degree(i)*50 for i in g.nodes()], width = [1 if g.degree(i)<= 5 else 3 for i in g.nodes()], node_color=[float(g.degree(i)) for i in g.nodes()],node_shape='+')  # Lo grafica delimita el tamao del nodo en la grafica
plt.show() #Lo muestra
```
