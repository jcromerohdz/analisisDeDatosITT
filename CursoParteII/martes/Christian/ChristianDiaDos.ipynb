## Agragar atributos

```python
nx.set_node_attributes(g, 'edad', 20)
nx.set_node_attributes(g, 'popularidad', g.degree())
```

```python
g.nodes('edad')
```

```python
g.nodes('popularidad')
```

```python
g[2]
```

```python
pos = nx.spring_layout(g)
```

```python
nx.draw(g, pos)
plt.show()
```

```python
pos = nx.circular_layout(g)
nx.draw(g, pos)
plt.show()
```

```python
pos = nx.random_layout(g)
nx.draw(g, pos)
plt.show()
```

```python
pos = nx.spectral_layout(g)
nx.draw(g, pos, node_size=1000, width=5)
plt.show()
```

```python
pos = nx.spring_layout(g)
nx.draw(g, pos, node_size=[g.degree(i) * 50 for i in g.nodes()])
plt.show()
```

```python
pos = nx.spring_layout(g)
nx.draw(g, pos, node_size=[g.degree(i) * 50 for i in g.nodes()], witdth =[1 if g.degree(i)<5 else 3 for i in g.nodes()],
       node_color=[float(g.degree(i)) for i in g.nodes()], node_shape="o")
plt.show()
```

```python
mg=nx.read_gml('mygraph.gml')
```

```python
nx.draw(mg)
plt.show()
```

```python
pos = nx.random_layout(mg)
nx.draw(mg, pos)
plt.show()
```

```python
pos = nx.spring_layout(mg)
nx.draw(mg, pos, node_size=[mg.degree(i) * 50 for i in mg.nodes()], witdth =[1 if mg.degree(i)<5 else 3 for i in mg.nodes()],
       node_color=[float(mg.degree(i)) for i in mg.nodes()])
plt.show()
```

```python
pos = nx.spring_layout(mg)
nx.draw(mg, pos, node_size=500)
plt.show()
```

```python
pos = nx.spring_layout(mg)
nx.draw(mg, pos, node_size=[mg.degree(i) * 100 for i in mg.nodes()], width =[1 if g.degree(i)<3 else 4 for i in g.nodes()],
       node_color=[float(mg.degree(i)) for i in mg.nodes()], with_labels=True, font_color="pink")
plt.show()
```
