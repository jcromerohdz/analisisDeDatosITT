```python
### AGREGAR ATRIBUTOS A LOS NODOS ###
```

```python
nx.set_node_attributes(g,'edad',20)
nx.set_node_attributes(g,'popularidad',g.degree())
```

```python
g.nodes('edad')
```

```python
pos =nx.spring_layout(g)
nx.draw(g,pos,node_size=[g.degree(i) * 50 for i in g.nodes()],width=[1 if g.degree(i)< 5 else 3 for i in g.nodes()],
       node_color = [float(g.degree(i))for i in g.nodes()], node_shape='o')
plt.show()
```

```python
g= nx. read_gml('mygraph.gml')
```
