```python
nx.set_node_attributes(g, 'edad',20) # Commando de todo hace el grafo
nx.set_node_attributes(g, 'popularidad', g.degree()) ## este varia
```

```python
pos = nx.spring_layout(g) # Ya tiene forma y no es alaeatoria la grafica @ SPRING
```

```python
nx.draw(g, pos)
plt.show()
```

```python
# FORMA CIRCULAR
# Depende el problema, por ejemplo ver la densidad o centralidad de los nodos, por ejemplo aqui vemos que estan a la derecha
# circular puede verse redes que estan centralizadas.

# Spring o Spectral, o escala libre es RANDOM
pos = nx.circular_layout(g)
nx.draw(g, pos)
plt.show()
```

```python
# FORMA RANDOM. es la DEFAULT

pos = nx.random_layout(g)
nx.draw(g, pos)
plt.show()
```

```python
# FORMA SPECTRAL
### MUESTRA SUBGRUPOS o Clustering, o aquellos que estan mas conectados (usado para conectar otros nodos) la influencia
pos = nx.spectral_layout(g)
nx.draw(g, pos)
plt.show()
```

```python
nx.draw(g, pos, node_size=1000, width=10) # un tipo de zoom
plt.show()
```

```python
pos = nx.spring_layout(g)
nx.draw(g, pos, node_size=[g.degree(i) * 50 for i in g.nodes()], witdth =[1 if g.degree(i)<5 else 3 for i in g.nodes()],
       node_color=[float(g.degree(i)) for i in g.nodes()], node_shape=">")
plt.show()
# no me funciono error de []
```
