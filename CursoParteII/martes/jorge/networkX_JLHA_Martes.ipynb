```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
g = nx.karate_club_graph()
```

```python
g.nodes()
```

```python
g.add_node(34)
```

```python
g.nodes()
```

```python
nx.draw(g, with_labels=True)
```

```python
plt.show()
```

```python
g.edges()
```

```python
g.add_edge(33,34)
```

```python
nx.draw(g,with_labels = True)
plt.show()
```

```python
g.add_edges_from([(34,26),(19,25),(10,18)])
```

```python
nx.draw(g,with_labels=True)
plt.show()
```

```python
g.edges(34)
```

```python
g.edges(30)
```

```python
nx.set_node_attributes(g,'edad',20)
nx.set_node_attributes(g, 'popularidad', g.degree())
```

```python
g.nodes('popularidad')
```

```python
pos = nx.spring_layout(g)
```

```python
nx.draw(g,pos)
```

```python
plt.show()
```

```python
posc=nx.circular_layout(g)
nx.draw(g,posc)
plt.show()
```

```python
posr = nx.random_layout(g)
nx.draw(g,posr)
plt.show()
```

```python
posr = nx.spectral_layout(g)
nx.draw(g,posr)
plt.show()
```

```python
posc=nx.spring_layout(g)
nx.draw(g,posc, node_size=1000, with_labels = True, width=3)
plt.show()
```

```python
posc=nx.spring_layout(g)
nx.draw(g,posc, node_size=[g.degree(i) * 50 for i in g.nodes()], with_labels = True, width=2)
plt.show()
```

```python
posc=nx.spring_layout(g)
nx.draw(g,posc, node_size=[g.degree(i) * 50 for i in g.nodes()], with_labels = True, width=[1 if g.degree(i)<5 else 3 for i in g.nodes()],
       node_color = [float(g.degree(i)) for i in g.nodes()])
plt.show()
```

```python
mg=nx.read_gml('mired.gml')
```
