# CIERRE DE ESTADISTICAS GRAFICAS

# matplotlib es la mas sencilla de las complejas


```python
import matplotlib.pyplot as plt
import numpy as np
# plot nos sirve para las series
n = np.random.uniform(-1,1,100)
m = np.random.uniform(-1,1,100)
#
plt.subplot(2,2,1)  # una micro figura, # Columnas, size 2 pulgadas, numero de la figura que va. 2=col,2=medida,3=orden
#
plt.plot(n, marker='x', color='r', label='f(x)') #r=rojo
plt.plot(m, marker='o', color='g', label ='f(y)')
plt.legend()
# Series y etiquetas
plt.xlabel('Tiempo')
plt.ylabel('Tipo de Cambio')
plt.title('Tipo de cambio historico del dolar')
plt.subplot(2,2,2) # antes del plot se pone

# Dispersa el rango mas pequeño, 
# para ver la relaciones y corelaciones existen.
plt.scatter(n,m)
plt.subplot(2,2,3) 

# Para la grafica de pae
labels = 'Frogs','Hogs', 'Dogs', 'Logs'
size = [15,30,45,10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0.1)
plt.pie(size, explode=explode, labels=labels, colors=colors,
       autopct='%1.1f%%', shadow=True, startangle=90) # '%1.1f%%' enteros con 1 decimal
plt.subplot(2,2,4)  # Antes de generar los plots va el muchas veces el sub
plt.plot(n) # con una sola vez
plt.show() # No lleva "n" Y ES SOLO UNA VEZ  AL FINAL
```
